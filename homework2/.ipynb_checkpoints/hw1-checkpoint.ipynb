```python
import pandas as pd
```

```python
# import data
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('ml-1m/users.dat', sep='::', header=None, names=unames, engine='python')

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('ml-1m/ratings.dat', sep='::', header=None, names=rnames, engine='python')

mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('ml-1m/movies.dat', sep='::', header=None, names=mnames, engine='python')
```

```python
data = pd.merge(pd.merge(ratings, users), movies)
```

```python
data[:5]
```

```python
# 1.An aggregate of the movie ratings for each particular genre.
mean_ratings_by_genres = data.pivot_table('rating', index='genres', aggfunc='mean')
```

```python
mean_ratings_by_genres
```

```python
mean_ratings_by_genres_gender = data.pivot_table('rating', index='genres',columns = 'gender',aggfunc='mean')
```

```python
mean_ratings_by_genres_gender[:5]
```

```python
# 2.The top 5 highest ranked genre by women.
top_female_ratings_by_genres = mean_ratings_by_genres_gender.sort_values(by='F', ascending=False)
```

```python
top_female_ratings_by_genres[:5]
```

```python
# 3.The top 5 highest ranked genre by men.
top_male_ratings_by_genres = mean_ratings_by_genres_gender.sort_values(by='M', ascending=False)
```

```python
top_male_ratings_by_genres[:5]
```

```python
# 5.A function that given a user_id and movie_id 
# returns a list of user_ids for other users
# that rated the movie identified by the provided movie_id with the same score
def sameUser(user_id, movie_id):
    selectedRating = data[(data.user_id == user_id) & (data.movie_id == movie_id)]['rating'].loc[0]
    sameMovieData = data[(data.movie_id == movie_id) & (data.rating == selectedRating)]
    removedDuplicateSameMovieData = sameMovieData.ix[sameMovieData['user_id']!=user_id]
    return removedDuplicateSameMovieData['user_id'].reset_index(level=0, drop=True)
```

```python
# example usage for sameUser function
sameUser(1,1193)[:10]
```

```python
# 6.additional topics:
# (1) Which movies do men and women most disagree on

data.reset_index('movie_id', inplace=True)
```

```python
pivoted = data.pivot_table(index=['movie_id', 'title'],
                           columns=['gender'],
                           values='rating',
                           fill_value=0)
```

```python
pivoted['diff'] = pivoted.M - pivoted.F 
```

```python
pivoted[:5]
```

```python
pivoted.reset_index('movie_id', inplace=True)
```

```python
most_40 = data.groupby('movie_id').size().sort_values(ascending=False)[:40]
```

```python
disagreements = pivoted[pivoted.movie_id.isin(most_40)]['diff']
disagreements.sort_values().plot(kind='barh', figsize=[9, 15])
plt.title('Male vs. Female Avg. Ratings\n(Difference > 0 = Favored by Men)')
plt.ylabel('Title')
plt.xlabel('Average Rating Difference');
plt.show()
```

```python
# (3) Binning our users and compare ratings across age groups.
import numpy as np
labels = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59']
data['age_group'] = pd.cut(data.age, range(0, 61, 10), right=False, labels=labels)
data[['age', 'age_group']].drop_duplicates()
data.groupby('age_group').agg({'rating': [np.size, np.mean]})
# explanation: (1) the older people has high rating on films
#              (2) people between 10 - 39 years old are the majority.
```
